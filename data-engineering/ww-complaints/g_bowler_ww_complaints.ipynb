```sql
%sql
CREATE OR REPLACE VIEW sts_prd.`01_ipd_thering`.g_bowler_ww_complaints AS
WITH product_scope AS (
    SELECT DISTINCT Product
    FROM hub_prd.g_bpc_finance.v_reltioproduct_hier
    WHERE Level1Desc IN (
        'Alaris System',
        'Hazardous Drug Solutions',
        'IV Solutions',
        'Gravity and Syringe Delivery',
        'IV Access',
        'CME',
        'Alaris Plus',
        'IPD OEM',
        'BD Nexus'
    )
    AND Product IS NOT NULL
)

-- regular table = master schema
SELECT
    c.Source,
    c.PrId,
    c.UniqueGridId,
    c.PrName,
    c.PrStatusType,
    c.PrDateOpened,
    c.PrDateClosed,
    c.PrIsClosed,
    c.PrDateCreated,
    c.ComplaintDateReceived,
    c.OrigDateClosed,
    c.AssignedTo,
    c.SuspectOrConcomitant,
    c.BusinessUnit,
    c.DisplayBu,
    c.DimSegment,
    c.CustomerName,
    c.AwarenessDate,
    c.RecType,
    c.Region,
    c.Country,
    c.MaterialValue,
    c.MaterialDesc,
    c.ServiceValue,
    c.ServiceComplaint,
    c.Investigation,
    c.InvestPrId,
    c.Dchu,
    c.ComplaintProductPlatform,
    c.ProductLevel1,
    c.ProductLevel2,
    c.RegionOfIntake,
    c.InitialReporterCity,
    c.PlantDescription,
    c.SerialNumber,
    c.BatchNumber,
    c.RecordState,
    c.ReOpenedOn,
    c.UsReportable,
    c.CaReportable,
    c.EuReportable,
    c.AsiaPacReportable,
    c.JapanReportable,
    c.LatamReportable,
    c.SelectIfPartOfLitigation,
    c.AsReportedCode1,
    c.AsReportedCode2,
    c.AsAnalyzedCode1,
    c.AsAnalyzedCode2,
    c.TypeOfInvestigationCode,
    c.InvestigationFinding,
    c.InvestigationConclusionCode,
    c.ComponentCodes,
    c.PatientProblemCodeAndTerms,
    c.AnnexECodeAndTerms,
    c.AnnexFCodeAndTerms,
    c.AnnexACodeAndTerms,
    c.AnnexACode,
    c.RowNum,
    c.ReportSource,
    c.ReportSourceOther,
    c.CustomerResponseNeeded,
    c.InitialReporterEMail,
    c.InitialReporter,
    c.EventOccurrenceDate,
    c.ExternalSystem,
    c.EmaReportable,
    c.CustomerType,
    c.UsMdrClassification,
    c.ComplaintCoordinator
FROM hub_prd.s_quality.semantic_complaintcounts_combined c
INNER JOIN product_scope p
    ON c.MaterialValue = p.Product

UNION ALL

SELECT
    h.SourceSystem                                                  AS Source,
    h.ComplaintId                                                   AS PrId,
    h.UniqueGridId                                                  AS UniqueGridId,
    h.ShortDescription                                              AS PrName,
    CAST(NULL AS INT)                                               AS PrStatusType,
    h.DateOpened                                                    AS PrDateOpened,
    h.DateClosed                                                    AS PrDateClosed,
    CAST(NULL AS INT)                                               AS PrIsClosed,
    CAST(NULL AS TIMESTAMP)                                         AS PrDateCreated,
    TO_TIMESTAMP(h.ComplaintDateReceived)                           AS ComplaintDateReceived,
    h.OrigDateClosed                                                AS OrigDateClosed,
    CAST(NULL AS STRING)                                            AS AssignedTo,
    h.SuspectOrConcomitant                                          AS SuspectOrConcomitant,
    h.BusinessUnit                                                  AS BusinessUnit,
    h.DisplayBusinessUnit                                           AS DisplayBu,
    h.MapSegment                                                    AS DimSegment,
    h.CustomerName                                                  AS CustomerName,
    h.AwarenessDate                                                 AS AwarenessDate,
    CAST(NULL AS STRING)                                            AS RecType,
    h.Region                                                        AS Region,
    h.EventOccurredCountry                                          AS Country,
    h.MaterialNumber                                                AS MaterialValue,
    h.MaterialDescription                                           AS MaterialDesc,
    CAST(NULL AS STRING)                                            AS ServiceValue,
    CAST(NULL AS STRING)                                            AS ServiceComplaint,
    h.InvestigationSummary                                          AS Investigation,
    CAST(NULL AS DECIMAL(38,0))                                     AS InvestPrId,
    h.Dchu                                                          AS Dchu,
    h.ProductPlatformOrFamily                                       AS ComplaintProductPlatform,
    h.ProductLevel1                                                 AS ProductLevel1,
    h.ProductLevel2                                                 AS ProductLevel2,
    h.RegionOfIntake                                                AS RegionOfIntake,
    h.CustomerCity                                                  AS InitialReporterCity,
    CAST(NULL AS STRING)                                            AS PlantDescription,
    h.SerialNo                                                      AS SerialNumber,
    h.BatchNo                                                       AS BatchNumber,
    h.RecordState                                                   AS RecordState,
    h.ReOpenedOn                                                    AS ReOpenedOn,
    CAST(NULL AS STRING)                                            AS UsReportable,
    CAST(NULL AS STRING)                                            AS CaReportable,
    CAST(NULL AS STRING)                                            AS EuReportable,
    CAST(NULL AS STRING)                                            AS AsiaPacReportable,
    CAST(NULL AS STRING)                                            AS JapanReportable,
    CAST(NULL AS STRING)                                            AS LatamReportable,
    h.SelectIfPartOfLitigation                                      AS SelectIfPartOfLitigation,
    h.AsReportedDeviceCodes                                         AS AsReportedCode1,
    CAST(NULL AS STRING)                                            AS AsReportedCode2,
    h.AsAnalyzedDeviceCodes                                         AS AsAnalyzedCode1,
    CAST(NULL AS STRING)                                            AS AsAnalyzedCode2,
    CAST(NULL AS STRING)                                            AS TypeOfInvestigationCode,
    CAST(NULL AS STRING)                                            AS InvestigationFinding,
    CAST(NULL AS STRING)                                            AS InvestigationConclusionCode,
    h.AsAnalyzedComponent                                           AS ComponentCodes,
    h.PatientProblemCodesAndTerms                                   AS PatientProblemCodeAndTerms,
    h.AnnexECode                                                    AS AnnexECodeAndTerms,
    h.AnnexFCode                                                    AS AnnexFCodeAndTerms,
    h.AnnexCCode                                                    AS AnnexACodeAndTerms,
    h.AnnexCCode                                                    AS AnnexACode,
    CAST(NULL AS STRING)                                            AS RowNum,
    CAST(NULL AS STRING)                                            AS ReportSource,
    CAST(NULL AS STRING)                                            AS ReportSourceOther,
    CAST(NULL AS STRING)                                            AS CustomerResponseNeeded,
    CAST(NULL AS STRING)                                            AS InitialReporterEMail,
    h.Originator                                                    AS InitialReporter,
    TO_TIMESTAMP(h.EventOccurrenceDate)                             AS EventOccurrenceDate,
    h.ExternalSystem                                                AS ExternalSystem,
    CAST(NULL AS STRING)                                            AS EmaReportable,
    CAST(NULL AS STRING)                                            AS CustomerType,
    CAST(NULL AS STRING)                                            AS UsMdrClassification,
    CAST(NULL AS STRING)                                            AS ComplaintCoordinator
FROM hub_prd.s_quality.semantic_complainthistory_combined h
INNER JOIN product_scope p
    ON h.MaterialNumber = p.Product;
```
