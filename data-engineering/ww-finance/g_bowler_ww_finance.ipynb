```sql
%sql
-- Actuals for IPD: Current Fiscal Year (Oct–Sep) AND Prior Fiscal Year
CREATE OR REPLACE VIEW sts_prd.`01_ipd_thering`.g_bowler_finance_actuals_ipd_cy AS
WITH fiscal_years AS (
  -- FY end year is determined by current date + 3 months (Oct–Sep fiscal)
  SELECT CAST(YEAR(ADD_MONTHS(CURRENT_DATE, 3)) AS STRING) AS fy_end_year  -- current FY
  UNION ALL
  SELECT CAST(YEAR(ADD_MONTHS(CURRENT_DATE, 3)) - 1 AS STRING)            -- prior FY
)
SELECT a.*
FROM hub_prd.g_bpc_finance.v_revenue_actual_mds_mms a
INNER JOIN hub_prd.g_bpc_finance.v_reltioproduct_hier b
    ON a.Reltio_Internal_ID = b.Reltio_Internal_ID
INNER JOIN fiscal_years fy
    ON SUBSTRING(a.FiscalYearPeriod, 1, 4) = fy.fy_end_year
WHERE b.Level1Desc IN (
    'Alaris System',
    'Hazardous Drug Solutions',
    'IV Solutions',
    'Gravity and Syringe Delivery',
    'IV Access',
    'CME',
    'Alaris Plus',
    'IPD OEM',
    'BD Nexus'
);
```

```sql
%sql
-- Querying Budget for IPD Current Fiscal Year (Oct–Sep) and saving as a view
CREATE OR REPLACE VIEW sts_prd.`01_ipd_thering`.g_bowler_finance_budget_ipd_cy AS
WITH current_fy AS (
  -- Fiscal year end year = current date + 3 months
  SELECT CAST(YEAR(ADD_MONTHS(CURRENT_DATE, 3)) AS STRING) AS fy_end_year
)
SELECT a.*
FROM hub_prd.g_bpc_finance.v_revenue_budget_mds_mms a
INNER JOIN hub_prd.g_bpc_finance.v_reltioproduct_hier b
    ON a.Reltio_Internal_ID = b.Reltio_Internal_ID
CROSS JOIN current_fy cf
WHERE b.Level1Desc IN (
    'Alaris System',
    'Hazardous Drug Solutions',
    'IV Solutions',
    'Gravity and Syringe Delivery',
    'IV Access',
    'CME',
    'Alaris Plus',
    'IPD OEM',
    'BD Nexus'
)
AND SUBSTRING(a.FiscalYearPeriod, 1, 4) = cf.fy_end_year;
```

```sql
%sql
-- Querying Projections for IPD Current Fiscal Year (Oct–Sep) and saving as a view
CREATE OR REPLACE VIEW sts_prd.`01_ipd_thering`.g_bowler_finance_projections_ipd_cy AS
WITH current_fy AS (
  SELECT CAST(YEAR(ADD_MONTHS(CURRENT_DATE, 3)) AS STRING) AS fy_end_year
)
SELECT a.*
FROM hub_prd.g_bpc_finance.v_revenue_projection_mds_mms a
INNER JOIN hub_prd.g_bpc_finance.v_reltioproduct_hier b
  ON a.Reltio_Internal_ID = b.Reltio_Internal_ID
CROSS JOIN current_fy cf
WHERE b.Level1Desc IN (
  'Alaris System',
  'Hazardous Drug Solutions',
  'IV Solutions',
  'Gravity and Syringe Delivery',
  'IV Access',
  'CME',
  'Alaris Plus',
  'IPD OEM',
  'BD Nexus'
)
AND SUBSTRING(a.FiscalYearPeriod, 1, 4) = cf.fy_end_year;
```
