```python
%sql
-- Create filtered, column-curated view for Sales Orders (IPD scope via Reltio categories)
CREATE OR REPLACE VIEW sts_prd.g_ipd.g_bowler_sales_order_all_ipd AS
WITH material_set AS (
  SELECT DISTINCT
    MaterialNumberHarmonized
  FROM hub_prd.s_isc.material_master_reltio
  WHERE ReltioLevel1Desc IN (
    'Alaris System',
    'Hazardous Drug Solutions',
    'IV Solutions',
    'Gravity and Syringe Delivery',
    'IV Access',
    'CME',
    'Alaris Plus',
    'IPD OEM',
    'BD Nexus'
  )
  AND MaterialNumberHarmonized IS NOT NULL
),
sales_orders AS (
  SELECT *
  FROM hub_prd.s_shared.sales_orders_all_sources
  WHERE MaterialNumberHarmonized IS NOT NULL
)
SELECT
    so.CreateDate,
    so.SourceSystem,
    so.OrderNumber,
    so.OrderType,
    so.ShipToNumber,
    so.SoldToNumber,
    so.MaterialNumber,
    so.MaterialNumberHarmonized,
    so.OrderStatusLineLast,
    so.OrderStatusLineNext,
    so.OrderStatusHeader,
    so.OrderStatusLineHarmonized,
    so.OrderStatusHeaderHarmonized,
    so.SalesDistributionChannel,
    so.SalesDistChannelHarmonized,
    so.ShipToName,
    so.SoldToName,
    so.Country,
    so.CustomerPo,
    so.OrderUom,
    so.BaseUom,
    so.ConversionFactor,
    so.OrderQuantityOriginal,
    so.OrderQuantityBase,
    so.CancelledQuantityOriginal,
    so.CancelledQuantityBase,
    so.DeliveredQuantityOriginal,
    so.OpenQuantityOriginal,
    so.OpenQuantityBase,
    so.DeliveredQuantityBase,
    so.CurrencyType,
    so.ExchangeRate,
    so.ExchangeDate,
    so.TotalPriceLocal,
    so.TotalPriceUsd,
    so.UnitPriceLocal,
    so.UnitPriceUsd,
    so.PromisedDeliveryDate,
    so.RequestedDeliveryDate,
    so.ActualShipDate,
    so.LastChangeDate,
    so.ExtractDate,
    so.BackOrderQuantityOriginal,
    so.BackOrderQuantityBase,
    so.LocalCurrency,
    so.ErpCustomerKey,
    so.AgreementNumber,
    so.ContractStartDate,
    so.ContractEndDate
FROM sales_orders so
INNER JOIN material_set ms
  ON so.MaterialNumberHarmonized = ms.MaterialNumberHarmonized;
```
